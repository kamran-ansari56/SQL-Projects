```python
engine = create_engine('sqlite:///yelp.db')
```

```python
def load__dataframe(df, table_name, engine):
    df.to_sql(table_name, con = engine, if_exists = 'replace', index = False)
```

```python
def load_dataframe_in_chunks(json_file, table_name, engine, chunksize=5000):
    """Load JSON file into SQL DB in chunks."""
    with pd.read_json(json_file, lines=True, chunksize=chunksize) as reader:
        for i, chunk in enumerate(reader):
            # Convert nested dict columns to JSON strings
            for col in ["attributes", "hours"]:
                if col in chunk.columns:
                    chunk[col] = chunk[col].apply(
                        lambda x: json.dumps(x) if isinstance(x, dict) else x
                    )
            
            # Write to SQL
            chunk.to_sql(table_name, engine, if_exists="append", index=False)
            print(f"Inserted chunk {i+1}")
```

```python
from sqlalchemy import text

with engine.connect() as conn:
    conn.execute(text("DROP TABLE IF EXISTS business;"))
    conn.execute(text("DROP TABLE IF EXISTS review;"))
    conn.execute(text("DROP TABLE IF EXISTS user;"))
    conn.execute(text("DROP TABLE IF EXISTS tip;"))
    conn.execute(text("DROP TABLE IF EXISTS checkin;"))
```

```python
load_dataframe_in_chunks("yelp_academic_dataset_business.json", "business", engine)
```

```python
load_dataframe_in_chunks("yelp_academic_dataset_review.json", "review", engine)
load_dataframe_in_chunks("yelp_academic_dataset_user.json", "user", engine)
load_dataframe_in_chunks("yelp_academic_dataset_tip.json", "tip", engine)
load_dataframe_in_chunks("yelp_academic_dataset_checkin.json", "checkin", engine)
```
